```python
def coincidencias_autor_categoriaMedicina():
    import pandas as pd
    df = pd.read_csv('ResultadoCategoriaFinal.csv')
    df_palabras_medicine = pd.read_csv('palabras.csv')
    
    #Permite separar por key y guardar su informacion.
    aux=0
    dicTitulos={}
    listautores=[]
    try:
        for i in df.index:
            aux=df['key'][i]
            aux1=df['key'][i+1]
            if aux==aux1:
                listautores.append(df['authors'][i])
                dicTitulos[aux]=listautores
            else:
                listautores=[]
    except:
        print('Fin')
     
    #Coincidencias
    dic_coincidencias={}
    list_titulos=list(dicTitulos.keys())
    for i in range(len(list_titulos)):
        #print(list_titulos[i])
        valor=str(dicTitulos.get(list_titulos[i]))
        valor=valor.replace("[", "") 
        valor=valor.replace("]", "") 
        valor=valor.replace('"', "") 
        valor=valor.replace("'", "") 
        valor=valor.split(',')
        #print(valor)
        frecuenciaPalab = []
        for w in valor:
            frecuenciaPalab.append(valor.count(w))
        #print("Pares\n" + str(list(zip(valor, frecuenciaPalab))))
        dic_coincidencias[list(dicTitulos.keys())[i]]=str(list(zip(valor, frecuenciaPalab)))
    return dic_coincidencias
    
```

```python
def autor_cetegoriaMedicina(dic_coincidencias):

    list_coincidencias=list(dic_coincidencias.keys())
    for i in range(len(list_coincidencias)):
        print(list_coincidencias[i])
        cate_medicina=list_coincidencias[i]
        print("------------------")
        nivel1=dic_coincidencias.get(list_coincidencias[i])
        nivel2=eval(nivel1)
        for i in range(len(nivel2)):
            #if para no buscar los espacios
            if nivel2[i][0] !=" ": 

                #print('sin:'+nivel2[i][0])
                #if de eliminacion de espacios
                if nivel2[i][0].find(' ')==0:
                    nombre=nivel2[i][0].replace(' ', '', 1)
                    if nombre !='none':
                        peso=nivel2[i][1]
                        relacion_neo4j_autor_categoriaMedicina(nombre,peso,cate_medicina)
                else:
                    nombre=nivel2[i][0]
                    peso=nivel2[i][1]
                    relacion_neo4j_autor_categoriaMedicina(nombre,peso,cate_medicina)
```

```python
def relacion_neo4j_autor_categoriaMedicina(nombre,peso,categoriaMedicina):
    print("Relacionar el autor:"+nombre+" con el peso :"+str(peso)+" en la categoria de medicina :"+categoriaMedicina)

    from py2neo import Graph
    graph = Graph("bolt://localhost:7687", auth=("neo4j", "123456789"))
    import re
    #Cargamos los esquemas de medicina y vivo

    try:
        graph.run('merge(valor1:persona{nombre:"'+nombre+'"})'
                     +'merge(valor2:owl__Class{rdfs__label:"'+categoriaMedicina+'"})'
                     +'merge (valor1)-[:rdfs__subClassOf{weight:'+str(peso)+'}]->(valor2)'  
                    );
    except:
        print("Error en :"+nombre)
```
