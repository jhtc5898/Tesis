```python
import pandas as pd
df_categoria_titulo = pd.read_csv('ResultadoCategoriaFinal.csv')
df_palabras_medicine = pd.read_csv('palabras.csv')
```

```python
list_categoria_titulo=df_categoria_titulo['categoriatitulo']
list_palabras_medicine=df_palabras_medicine["n.rdfs__label"]
```

```python
list_posicion_medicina=[]
list_posicion_categoria=[]
list_titulo_medicina=[]
list_titulo_articulo=[]
list_peso=[]

palabrasmal=[]
for i in range(10):
    print(i)
    print(list_palabras_medicine[i])
    #print("-------------------")
    titulo_medicine=list_palabras_medicine[i]
    titulo_medicine=titulo_medicine.lower()
    for j in range(len(list_categoria_titulo)):
        try:
            categoria=list_categoria_titulo[j]
            output=eval(categoria)
            output=list(set(output))
            #print(output)
            peso=1/len(output)
            aux =0
            for x in range(len(output)):
                categoria_palabra=output[x].split(',')
                #print(categoria_palabra[0])
                palabra=categoria_palabra[0]
                palabra=palabra.lower()
                if titulo_medicine.find(palabra)>=0:
                    aux=aux+peso
                    #print("en el ,",titulo_medicine,'se econtro la palabra',palabra)
            
            #print("tiene un peso:",aux)
            if aux>0:
                list_posicion_medicina.append(i)
                list_posicion_categoria.append(j)
                list_titulo_medicina.append(titulo_medicine)
                list_titulo_articulo.append(df_categoria_titulo["title"][j])
                list_peso.append(float(aux)*100)
                print("tiene un peso:",aux)
                
                
        except:
            palabrasmal.append([i,j])
    
```

```python
import pandas as pd  
dic_general ={'posicion_medicina':list_posicion_medicina ,'posicion_categoria':list_posicion_categoria,'titulo_medicina':list_titulo_medicina,'titulo_articulo':list_titulo_articulo,'peso':list_peso}
dataset=pd.DataFrame(dic_general)
print(dataset)
dataset.to_csv('CategoriaMedicinaPesos.csv')
```
